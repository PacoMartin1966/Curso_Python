**1) Ejercicio de herencia múltiple y polimorfismo:**
* Crea la clase Vehiculos con un constructor que incluya marca y modelo
* Esta clase Vehiculos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar gasolina"
* Crea la clase VElectricos con un constructor que incluya marca, modelo y autonomia
* Esta clase VElectricos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar electricidad"
* Crea la clase BicicletaElectrica que herede de Vehiculos y de VElectricos, pero dando prioridad a VElectricos (ya que es un vehículo más electrico que normal)
* Crea la clase Quad que herede de Vehiculos y de VElectricos, pero dando prioridad a Vehiculos (ya que sólo usa la electricidad de modo puntual)
* Crear un objeto de BicicletaElectrica y otro de Quad
* Emplear las técnicas de polimorfismo aprendidas para conseguir que al peguntarle a los objetos que acabamos de crear por su repostaje, ambos nos respondan adecuadamente



```python
# Completa el ejercicio aquí
class Vehiculos:
    """Cración de la clase Vehículos, la fuente de energía es la gasolina"""
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
    
    def repostar(self):
        print("Este vehiculo tiene que repostar gasolina")
        
class VElectricos:
    """Cración de la clase Vehículos Eléctricos, la fuente de energía es la electricidad"""
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        self.autonomia = 100
    
    def repostar(self):
        print("Este vehiculo tiene que repostar electricidad")
        
class BicicletaElectrica(VElectricos, Vehiculos):
    """Cración de la clase Bicicletas eléctricas, heredan principalmente de vehículos eléctricos y de vehículos"""
    pass

class Quad(Vehiculos, VElectricos):
    """Cración de la clase Quad, heredan principalmente de vehículos y de vehículos eléctricos"""
    pass

def repostar_vehiculo(vehiculo):
    print(vehiculo.repostar())

# Creamos dos objetos
Bicicleta_1 = BicicletaElectrica("Orbea", "ST200")
Quad_1 = Quad("Toyota", "TY250")

# Imprimimos por pantalla la funcion repostar de cada uno.
print(Bicicleta_1.repostar())
print(Quad_1.repostar())

# Polimorfismo
repostar_vehiculo(Bicicleta_1)
repostar_vehiculo(Quad_1)
```

```python
# Completa el ejercicio aquí
class Personal_Universitario:

    def __init__(self, ide, nombre, email):

        self.diccionario = {"id":ide, "nombre":nombre,"email":email} # Se inicializa el diccionario

        

    def descripcion(self): # Se imprimen los datos del diccionario

        print("Los datos del objeto {} son:\nId: {} \nNombre: {} \nEmail: {}".format(type(self).__name__,
                                                                                     self.diccionario["id"],
                                                                                     self.diccionario["nombre"],
                                                                                     self.diccionario["email"]))
    
        

class Oficina(Personal_Universitario):
    """Clase Oficina hereda de Personal_Universitario"""
    def __init__(self,ide, nombre, email, puesto):
        super().__init__(ide, nombre, email)
        self.puesto = puesto
        self.diccionario['puesto'] = self.puesto
        
    def descripcion(self): # Se imprimen los datos del diccionario

        print("Los datos del objeto {} son:\nId: {} \nNombre: {} \nEmail: {} \nPuesto {}:".format(type(self).__name__,
                                                                                                  self.diccionario["id"],
                                                                                                  self.diccionario["nombre"],
                                                                                                  self.diccionario["email"], 
                                                                                                  self.diccionario["puesto"]))
        
    

        
        
class Profesor(Personal_Universitario):
    """Clase Oficina hereda de Personal_Universitario"""
    def __init__(self,ide, nombre, email, especializacion):
        super().__init__(ide, nombre, email)
        self.especializacion = especializacion
        self.diccionario['especializacion'] = self.especializacion
        
    def descripcion(self): # Se imprimen los datos del diccionario

        print("Los datos del objeto {} son:\nId: {} \nNombre: {} \nEmail: {} \nEspecializacion {}:".format(type(self).__name__,
                                                                                                           self.diccionario["id"],
                                                                                                           self.diccionario["nombre"],
                                                                                                           self.diccionario["email"],
                                                                                                           self.diccionario["especializacion"]))


class Alumno(Personal_Universitario):
    """Clase Oficina hereda de Personal_Universitario"""
    def __init__(self,ide, nombre, email, creditos_aprobados):
        super().__init__(ide, nombre, email)
        self.creditos_aprobados = creditos_aprobados
        self.diccionario['creditos_aprobados'] = self.creditos_aprobados
        
    def descripcion(self): # Se imprimen los datos del diccionario

        print("Los datos del objeto {} son:\nId: {} \nNombre: {} \nEmail: {} \nCreditos Aprobados: {}".format(type(self).__name__,self.diccionario["id"],self.diccionario["nombre"],self.diccionario["email"], self.diccionario["creditos_aprobados"]))

```

```python
class Supermercado:
    def __init__(self, ide, nombre, pvp, ubicacion):
        self.ide = ide
        self.nombre = nombre
        self.pvp = pvp
        self.ubicacion = ubicacion
        
    def __str__(self):
        return """\
        FAMILIA PRODUCTO\t{}
        IDE\t\t{}
        NOMBRE\t\t{}
        PVP\t\t{}
        UBICACION\t{}""".format(type(self).__name__,self.ide, self.ide, self.nombre, self.pvp, self.ubicacion)

class Bebidas(Supermercado):
    def __init__(self, ide, nombre, pvp, ubicacion, distribuidor):
        super().__init__(ide, nombre, pvp, ubicacion)
        self.distribuidor = distribuidor
        
    def __str__(self):
        return """\
        FAMILIA PRODUCTO\t{}
        IDE\t\t\t{}
        NOMBRE\t\t\t{}
        PVP\t\t\t{}
        UBICACION\t\t{}
        DISTRIBUIDOR\t\t{}""".format(type(self).__name__,self.ide, self.nombre, self.pvp, self.ubicacion, self.distribuidor)

class Alcoholicas(Bebidas):
    pass

class Cervezas(Alcoholicas):
    pass

class Vinos(Alcoholicas):
    pass


class Aguas(Bebidas):
    pass

class Ferreteria(Supermercado):
    def __init__(self, ide, nombre, pvp, ubicacion, fabricante):
        super().__init__(ide, nombre, pvp, ubicacion)
        self.fabricante = fabricante
        
    def __str__(self):
        return """\
        FAMILIA PRODUCTO\t{}
        IDE\t\t\t{}
        NOMBRE\t\t\t{}
        PVP\t\t\t{}
        UBICACION\t\t{}
        FABRICANTE\t\t{}""".format(type(self).__name__,self.ide, self.nombre, self.pvp, self.ubicacion, self.fabricante)

class Tornillos(Ferreteria):
    pass

class Electrico(Ferreteria):
    pass

class Enchufes(Electrico):
    def __init__(self, ide, nombre, pvp, ubicacion, fabricante, tension_trabajo):
        super().__init__(ide, nombre, pvp, ubicacion,fabricante)
        self.tension_trabajo = tension_trabajo
    
    def __str__(self):
        return """\
        FAMILIA PRODUCTO\t{}
        IDE\t\t\t{}
        NOMBRE\t\t\t{}
        PVP\t\t\t{}
        UBICACION\t\t{}
        FABRICANTE\t\t{}
        TENSION DE TRABAJO\t{}""".format(type(self).__name__,self.ide, self.nombre, self.pvp, self.ubicacion, self.fabricante, self.tension_trabajo)

# ide, nombre, pvp, ubicacion

#botella = Supermercado('0001', 'Botella 1L', 12, 'A1')
lista = []
Lanjaron = Aguas('0001', 'Botella 1L Lanjaron', 0.55, 'A1', 'Mercadona')
lista.append(Lanjaron)
Montseny = Aguas('0002', 'Botella 1L Montseny', 0.45, 'B1', 'Carrefour')
lista.append(Montseny)
MarquesCaceres = Vinos('0003', 'Botella 75cl Marques de Cáceres', 6.60, 'C2', 'Marques de Cáceres')
lista.append(MarquesCaceres)
Mahou = Cervezas('0004', 'Botella 33cl Mahou', 1.25, 'C2', 'Mahou')
lista.append(Mahou)
Estrella = Cervezas('0025', 'Botella 20cl Estrella', 1.30, 'C3', 'Damm')
lista.append(Estrella)
#ide, self.nombre, self.pvp, self.ubicacion, self.fabricante
M3 = Tornillos('0005', 'Tornillo M-3', 0.2, 'F3', 'Micalor')
lista.append(M3)
M4 = Tornillos('0006', 'Tornillo M-4', 0.3, 'F4', 'MecTrun')
lista.append(M4)
# ide, nombre, pvp, ubicacion, fabricante, tension_trabajo
Basenormal = Enchufes('0007', 'Base de enchufe normal', 5.60, 'J3', 'Simon', '240V')
lista.append(Basenormal)
BaseSchuko = Enchufes('0008', 'Base de enchufe Schuko', 9.20, 'J7','Legrand', '230V')
lista.append(BaseSchuko)
BombillaLed = Electrico('0009', 'Bombilla led', 12.90, 'J9', 'Philips')
lista.append(BombillaLed)
```

```python
def subida_precio(p, incremento):
    """Subida precio producto en porcentaje"""
    p.pvp = round(p.pvp * incremento, 2)
    
print("LISTADO DE TODOS LOS PRODUCTOS\n")    
for producto in lista:
    print(producto, "\n")
print("\n\nLISTADO DE SUBIDAS DE TODOS LOS PRODUCTOS\n")    

for producto in lista:    
    print("ANTES DE LA SUBIDA: \t", end="")
    print(producto.nombre, "\t", producto.pvp)
    subida_precio(producto, 1.25) # SUbida de los productos del 25%
    print("DESPUES DE LA SUBIDA: \t", end="")
    print(producto.nombre, "\t", producto.pvp)
    print()
    
print("LISTADO DE TODOS LOS PRODUCTOS DESPUES DE LA SUBIDA\n")    
for producto in lista:
    print(producto, "\n")
```
