```python
# Completa el ejercicio aquí

# importamos la librería
import sqlite3
```

```python
# conectamos con la base de datos
conexion = sqlite3.connect("biblioteca.db")

# creamos las tablas

# primero creamos el cursor
cursor = conexion.cursor()

# creamos las tablas
cursor.execute("CREATE TABLE autor (dni VARCHAR(9), nombre VARCHAR(50) NOT NULL, apellidos VARCHAR(100) NOT NULL, estarVivo BOOLEAN NOT NULL)")
cursor.execute("CREATE TABLE libro (isbn VARCHAR(13), titulo VARCHAR(300) NOT NULL, editorial VARCHAR(50) NOT NULL, año_escrito SMALLINT)")
cursor.execute("CREATE TABLE usuario (dni VARCHAR(9), nombre VARCHAR(30) NOT NULL, apellidos VARCHAR(30) NOT NULL, numPrestamos SMALLINT DEFAULT(0))")

# guardamos cambios y escribimos en la base de datos
conexion.commit()

# cerramos la conexion
conexion.close()
```

```python
import sqlite3

conexion = sqlite3.connect("biblioteca.db")
cursor = conexion.cursor()

cursor.execute("SELECT * FROM autor")

autores = cursor.fetchall()

print("\tAUTORES\n")
for autor in autores:
    print(autor[1], autor[2] )

cursor.execute("SELECT * FROM libro")

libros = cursor.fetchall()

print("\n\tLIBROS\n")
for libro in libros:
    print(libro[1])
    
cursor.execute("SELECT * FROM usuario")

usuarios = cursor.fetchall()

print("\n\tUSUARIOS\n")
for usuario in usuarios:
    print(usuario[1],usuario[2])
    
cursor.execute("SELECT * FROM autor WHERE estarVivo = True")

autores = cursor.fetchall()

print("\n\tAUTORES VIVOS\n")
for autor in autores:
    print(autor[1],autor[2])
    
cursor.execute("SELECT * FROM libro WHERE año_escrito > 1900")

libros = cursor.fetchall()

print("\n\tLIBROS ESCRITO POSTERIOR A 1900\n")
for libro in libros:
    print(libro[1],libro[2])
    
cursor.execute("SELECT * FROM usuario WHERE numPrestamos > 10 AND nombre = 'Paco'")

usuarios = cursor.fetchall()

print("\n\tUSUARIOS QUE SE HAYAN LLEVADO MAS DE 10 LIBROS Y SE LLAMEN PACO\n")
for usuario in usuarios:
    print(usuario[1],usuario[2], usuario[3], "Libros ")
    
conexion.commit()
conexion.close()
```
