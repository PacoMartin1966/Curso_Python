**Ejercicio de puntos, coordenadas y vectores sobre el plano cartesiano. A continuación los conceptos básicos.**

### El plano cartesiano

Representa un espacio bidimensional (en 2 dimensiones), formado por dos rectas perpendiculares, una horizontal y otra vertical que se cortan en un punto. La recta horizontal se denomina eje de las abscisas o **eje X**, mientras que la vertical recibe el nombre de eje de las ordenadas o simplemente **eje Y**. En cuanto al punto donde se cortan, se conoce como el **punto de origen O**.
**El plano se divide en 4 cuadrantes.**

<img src="./Ejes.png"  />



### Puntos y coordenadas

El objetivo de todo esto es describir la posición de **puntos** sobre el plano en forma de **coordenadas**, que se forman asociando el valor del eje de las X (horizontal) con el valor del eje Y (vertical).

La representación de un punto es sencilla: **P(X,Y)** dónde X y la Y son la distancia horizontal (izquierda o derecha) y vertical (arriba o abajo) respectivamente, utilizando como referencia el punto de origen (0,0), justo en el centro del plano.

<img src="./Puntos.png"  />


### Vectores en el plano

Finalmente, un vector en el plano hace referencia a un segmento orientado, generado a partir de dos puntos distintos. 

A efectos prácticos no deja de ser una línea formada desde un punto inicial en dirección a otro punto final, por lo que se entiende que un vector tiene longitud y dirección/sentido.

<img src="./Vector.png"  />


En esta figura, podemos observar dos puntos A y B que podríamos definir de la siguiente forma:
* **A(x1, y1)** => **A(2, 3)**
* **B(x2, y2)** => **B(7, 5)**

Y el vector se representaría como la diferencia entre las coordenadas del segundo punto respecto al primero (el segundo menos el primero):
* **AB = (x2-x1, y2-y1)** => **(7-2, 5-3)** => **(5,2)** 

Lo que en definitiva no deja de ser: 5 a la derecha y 2 arriba.

````python
# Realizar el ejercicio aquí

import math

class Punto:
    
    def __init__(self, x=None, y=None):
        if x == None:
            x = 0
        if y == None:
            y = 0
        self.x = x
        self.y = y
        
    def __str__(self):
        return "({},{})".format(self.x, self.y)
    
    def cuadrante(self):
        x = self.x
        y = self.y
        
        if x == 0 and y == 0:
            print("El punto ({},{}) es el origen".format(x,y))
        elif x > 0 and y > 0:
            print("El punto ({},{}) está en el primer cuadrante".format(x,y))
        elif x < 0 and y > 0:
            print("El punto ({},{}) está en el segundo cuadrante".format(x,y))
        elif x < 0 and y < 0:
            print("El punto ({},{}) está en el tercer cuadrante".format(x,y))
        elif x > 0 and y < 0:
            print("El punto ({},{}) está en el cuarto cuadrante".format(x,y))
        else:
            print("Falta determinar otra opción")
            
    def vector(self, punto2):
        print("Estoy en vector", self.x)
        x2 = punto2.x
        y2 = punto2.y
        vector = Punto(x2 - self.x, y2 - self.y)
        return vector
    
    def distancia(self, punto2):
        x2 = punto2.x
        y2 = punto2.y
        distancia = round(math.sqrt((x2 - self.x)**2 + (y2 - self.y)**2))
        return distancia     
        
class Rectangulo:
    
    punto_inicial = []
    punto_final = []
    
    def __init__(self, inicial = None, final = None):
        
        self.punto_inicial = []
        self.punto_final = []
        
        if inicial == None:
            self.punto_inicial.append(0)
            self.punto_inicial.append(0)
            print("Inicial", self.punto_inicial)
        else:
            self.punto_inicial.append(inicial.x)
            self.punto_inicial.append(inicial.y)
            print("Inicial", self.punto_inicial)
            
        if final == None:
            self.punto_final.append(0)
            self.punto_final.append(0)
            print("Final", self.punto_final)
        else:
            self.punto_final.append(final.x)
            self.punto_final.append(final.y)
            print("Final", self.punto_final)
            
    def base(self):
         
        print("Estoy en base")
        punto1Base = [self.punto_inicial[0], self.punto_inicial[1]]
        punto2Base = [self.punto_final[0], self.punto_inicial[1]]
        print("Primer punto base", punto1Base)
        print("Segundo punto base", punto2Base)
        return punto1Base, punto2Base
    
    def altura(self):
        print("Estoy en altura")
        punto1Altura = [self.punto_final[0], self.punto_inicial[1]]
        punto2Altura = [self.punto_final[0], self.punto_final[1]]
        print("Primer punto altura", punto1Altura)
        print("Segundo punto altura", punto2Altura)
        return punto1Altura, punto2Altura
    
    def area(self, base, altura):
        
        area = base * altura / 2
        return area
        
           
````
